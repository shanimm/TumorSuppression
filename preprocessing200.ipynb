```python
df1=pd.read_csv('finalxyztrialnew.csv')
from sklearn.preprocessing import minmax_scale
df1.head()
```

```python
# minmaxing the coordinates
df1[['x_coord','y_coord','z_coord']]=minmax_scale(df1[['x_coord','y_coord','z_coord']])
```

```python
#mulitplying by 200 to get range of 0-200 throughout
df1[['x_coord','y_coord','z_coord']]=200*df1[['x_coord','y_coord','z_coord']]
```

```python
df_xy=df1.drop('z_coord',axis=1)
df_yz=df1.drop('x_coord',axis=1)
df_xz=df1.drop('y_coord',axis=1)
df_xyz=df1
df_xy.to_csv('xyPDBdatafinal1.csv')
df_yz.to_csv('yzPDBdatafinal1.csv')
df_xz.to_csv('xzPDBdatafinal1.csv')
df_xyz.to_csv('xyzPDBdatafinal1.csv')
```
